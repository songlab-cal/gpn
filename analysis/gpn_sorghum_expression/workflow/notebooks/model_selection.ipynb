```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
steps_per_epoch = 265

genes = [
    "PsbS",
    "SBPase",
    "Raf1",
]

checkpoints = [
    f"checkpoints_epoch/GPN_Brassicales/30_epochs/{seed}/checkpoint-{steps_per_epoch * i}"
    for i in range(1, 31)
    for seed in [
        42,
        43,
        44,
        45,
        46,
    ]
]
```

```python
res = []
for gene in genes:
    for checkpoint in checkpoints:
        df = pd.read_csv(f"../../results/experimental_data/metrics/{gene}/{checkpoint}.csv")
        df = df[df.subset=="all"]
        df["gene"] = gene
        df["checkpoint"] = checkpoint
        res.append(df)
res = pd.concat(res)
res["epoch"] = res.checkpoint.str.extract(r"checkpoint-(\d+)").astype(int) // steps_per_epoch
res["seed"] = res.checkpoint.str.split("/").str[-2]
res.drop(columns=["checkpoint", "subset"], inplace=True)
res
```

```python
res.to_csv("metrics_MPRA.csv.gz", index=False)
```
